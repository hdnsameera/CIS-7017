### Concatenate all DataFrames as df_All

```python
df_All = pd.concat([df_Sitting, df_Tennis, df_Football, df_Cycling, df_JumpRope, df_Swimming, df_WalkingUpstairs, df_WalkingDownstairs, df_Walking, df_Pushups, df_Jogging], axis=0)
df_All
```

```python
df_All.columns
```

```python
def fnMinMaxNormalize(x):
    return (x - min(x)) / (max(x) - min(x))
```

### Remove ACTIVITY column and timestamps
#### Apply normalization

```python
df_temp = df_All
df_All = df_All.drop(columns=['ACTIVITY', 'Time since start in ms ', 'YYYY-MO-DD HH-MI-SS_SSS']).apply(fnMinMaxNormalize)
```

```python
df_All
```

```python
# Assign Timestamp and ACTIVITY
df_All['TIME_SINCE_START'] = df_temp['Time since start in ms ']
df_All['YYYYMMDD_HHMISS_SSS'] = df_temp['YYYY-MO-DD HH-MI-SS_SSS']
df_All['ACTIVITY'] = df_temp['ACTIVITY']
df_All
```

#### Number of observations of each activity extracted from the full dataframe

```python
activity = df_All.ACTIVITY.unique()
count = 0

while count < len(activity):
    
    print(activity[count] + " - " + str(len(df_All[df_All['ACTIVITY']== activity[count]])))
    count += 1
```
