```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

#### Listing down all the CSV files in the 'data' directory

```python
# The 'data' directory contains only the datasets of the activities

import os

file = []
for dirname, _, filenames in os.walk('data'):
    for filename in filenames:
        if ".csv" in filename:
            file.append(os.path.join(filename))
print(file)
print('\n Number of files = '+str(len(file)))
```

#### Generating DataFrames and reading CSV files dynamically

```python
i = 0
for f in file:
    
    # Removing the sufix '.csv' part in order to extract the name of the activity from the file name.
    f = f.replace('.csv', '')
    
    # Creating names of the dataframes
    df = 'df_' + str(i)
    
    # Reading CSV files dynamically in to dynamically created dataframes
    globals()[df] = pd.read_csv('data/'+file[i])
    
    # Adding 'ACTIVITY' column and assigning the activity name extracted from the file name.
    globals()[df]['ACTIVITY'] = f.upper()
    
    # Printing column names of all the dataframes
    print(globals()[df].columns)
    
    # Printing the number of columns of each dataframe
    print('Number of columns = '+ str(len(globals()[df].columns)) + '\n')
    
    # Incrementing the index of the loop
    i += 1    
```

#### Normalization function

```python
def fnMinMaxNormalize(x):
    return (x - min(x)) / (max(x) - min(x))
```
